# 1. Prepare Toronto data file


The postal codes of Toronto (starting with M) will be scraped from <a href="https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M">the List of postal codes of Canada: M</a>.

```python
#read the list from the website
import pandas as pd
html = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
html
    
```

```python
#Check how many tables read in the list
len(html)
```

```python
#Find the right table
html[0]
```

```python
#Organize the table
table = pd.DataFrame(html[0])
print(table.head())
print(table.shape)
```

```python
#Move the first row up as the header
table.columns = table.iloc[0]
table = table[1:]
table.rename(columns={'Postcode':'PostalCode'}, inplace=True)
print(table.head())
print(table.shape)
```

```python
#Reset index
table.reset_index(drop=True, inplace=True)
print(table.head())
print(table.shape)
```

```python
#Clean up boroughs are Not assigned
data_tor = table
data_tor.drop(data_tor[data_tor['Borough'] == 'Not assigned'].index, inplace=True)
data_tor.reset_index(drop=True, inplace=True)
print(data_tor.head())
print(data_tor.shape)
```

```python
#Change not assigned neighbourhood to the value of the borough.
data_tor.loc[data_tor.Neighbourhood =='Not assigned', 'Neighbourhood'] = data_tor.Borough
#Check if any neighbourhood signed as "Not assigned"
print(data_tor[data_tor['Neighbourhood']=='Not assigned'])
print(data_tor.head())
print(data_tor.shape)
```

```python
#Combine the values of neighbourhood with same borough
data_tor_p1 = data_tor.groupby(['PostalCode','Borough'])['Neighbourhood'].agg(lambda c: ', '.join(c)).reset_index()
data_tor_p1
```

```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

3) Explore the neighbourhoods in the Beaches.

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            limit)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
theBeaches_venues = getNearbyVenues(names=tor_data['Neighbourhood'],
                                   latitudes=tor_data['Latitude'],
                                   longitudes=tor_data['Longitude']
                                  )
```

4) Analyze Each Neighborhood

```python
# one hot encoding
theBeaches_onehot = pd.get_dummies(theBeaches_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
columns = theBeaches_venues['Neighborhood']
theBeaches_onehot['Neighborhood'] = columns 

#find index of column "Neighborhood"
index_neighborhood = theBeaches_onehot.columns.get_loc("Neighborhood")

# move neighborhood column to the first column
fixed_columns = [theBeaches_onehot.columns[index_neighborhood]] + list(theBeaches_onehot.columns[:index_neighborhood])
theBeaches_onehot = theBeaches_onehot[fixed_columns]

theBeaches_onehot.head(8)
```

```python
theBeaches_onehot.shape
```

```python
#group rows by neighborhood and by taking the mean of the frequency of occurrence of each category
theBeaches_grouped = theBeaches_onehot.groupby('Neighborhood').mean().reset_index()
theBeaches_grouped
```

```python
theBeaches_grouped.shape
```

```python
#print each neighborhood along with the top 5 most common venues

num_top_venues = 5

for hood in theBeaches_grouped['Neighborhood']:
    print("----"+hood+"----")
    temp = theBeaches_grouped[theBeaches_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
#Put the above result into panda dataframe. Write a function to sort the venues in descending order.
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
#create the new dataframe and display the top 10 venues for each neighborhood

num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = theBeaches_grouped['Neighborhood']

for ind in np.arange(theBeaches_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(theBeaches_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

5) Cluster Neighborhoods

```python
#Run k-means to cluster the neighborhood into 5 clusters.
# set number of clusters
kclusters = 5

theBeaches_grouped_clustering = theBeaches_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(theBeaches_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
#create a new dataframe that includes the cluster as well as the top 10 venues for each neighborhood.
# add clustering labels
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

theBeaches_merged = tor_data

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
theBeaches_merged = theBeaches_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

theBeaches_merged.head() # check the last columns!
```

```python
# visualize the resulting clusters

# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(theBeaches_merged['Latitude'], theBeaches_merged['Longitude'], theBeaches_merged['Neighborhood'], theBeaches_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
